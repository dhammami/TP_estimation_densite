```python
# -*- coding: utf-8 -*-

## Exercice 1
    ## 1
        ## a
from random import gauss

rg = range(50)

x1 = [gauss(-1.5, 2) for i in rg]
print ("Valeur de x1 :")
print (x1)

        ## b
from random import uniform

x2 = [uniform(0, 1) for i in rg]
print ("Valeur de x2 :")
print (x2)

        ## c
import numpy as np

X = np.concatenate((x1, x2), axis=0)
print ("\nValeure de X obtenu :")
print (X)

    ## 2
        ## a
import matplotlib.pyplot as plt

plt.hist(X, normed=1)
plt.show()

        ## b
import math

def K_gaussien(u):
    return (1.0 / math.sqrt(2 * math.pi)) * math.exp(-(u ** 2) / 2.0)


def K_silverman(u):
    return (1.0 / 2.0) * math.exp((-abs(u) * 1.0) / math.sqrt(2)) * sin(((abs(u) * 1.0) / math.sqrt(2)) + math.pi / 4.0)


print("\nExemple de r√©sultat pour le noyau Gaussien : " + str(K_gaussien(2)))

def f (h, X, x, K) :
    """
    Fonction : f
    Parametres : - Une valeur de h petite
                 - Le jeu de donn√©es
                 - Une valeur pour x pour laquelle on calcule la fonction
                 - Une fonction qui servira de noyau
    Retour : la valeur de l'estimateur de la fonction calcul√© √† x avec les autres param√®tres
    Description : Calcule la valeur de l'estimateur de la densit√© calcul√© en x selon la methode des noyaux
    """
    f = 0
    for i in X :
        f += K (((i - x) * 1.0) / h)
    return 1.0 / (len(X) * h) * f

## On choisit h = 0.25
h = 0.25
x = np.linspace(start=-10, stop=10, num=200, endpoint="True")
y = np.linspace(start=0, stop=len(x), num=len(x))

i = 0
while i < len(x):
    y[i] = f(h, X, x[i], K_gaussien)
    i += 1

plt.plot(x, y, color="blue")
plt.show()


        ## c
def f_m_i(x, h, i, X, K):
    """
    Fonction : f_m_i
    Parametres : - la valeur √† laquelle estimer la fonction
                 - une valeur de h petite
                 - un rang √† ne pas prendre en compte parmis le jeu de donn√©es
                 - un jeu de donn√©es
                 - une fonction noyau √† utiliser
    Retour : la valeur de f^(-i)
    Calcul : retourne la valeur de f^(-i)
    """
    somme = 0
    for j in range(len(X)):
        if j != i:
            somme += K_gaussien((X[j] - x) / h)

    return (1 / ((len(X) - 1) * h)) * somme


def f_carre(x, h, X, K):
    return f (h, X, x, K) ** 2


import scipy.integrate as integrate


def J (h, X, K):
    somme = 0
    for i in range(len(X)) :
        somme += f_m_i (X[i], h, i, X, K)
    return integrate.quad (
        f_carre,
        -np.inf,
        np.inf,
        args=(h, X, K))[0] - ((2 / len(X)) * somme
    )


print("Valeur de J pour le h utilis√© pr√©c√©dement : " + str(J(h, X, K_gaussien)))

def trace_f_avec_h_optimise (X, col) :
    """
    Fonction : trace_f_avec_h_optimise
    Parametre : - un jeu de donn√©es dont on veut estimer la densit√©
                - la couleur des graphiques √† l'affichage
    Retour : aucun
    Description : - trouve une valeur de h optimale avec la methode de validation
        crois√©e et l'affiche
                  - affiche la courbe des erreurs pour v√©rifier que le h trouv√©
        en est bien le minimum
                  - affiche la courbe de la densit√© ainsi obtenue
    """

    n = len(X) * 1.0

    C = np.linspace(start=-1, stop=n ** (1.0 / 3.0), num=100, endpoint="True")

    h_opti = 1
    J_h_opti = J(h_opti, X, K_gaussien)

    h_aff = np.linspace(start=0, stop=1, num=100)
    J_aff = np.linspace(start=0, stop=1, num=100)

    i = 0
    for c in C:
        h_test = c * (n ** -(1.0 / 3))
        h_aff[i] = h_test
        J_h_test = J(h_test, X, K_gaussien)
        J_aff[i] = J_h_test
        i += 1

        if J_h_test < J_h_opti:
            h_opti = h_test
            J_h_opti = J_h_test

    print("h optimum trouv√© : " + str(h_opti))

    # On affiche la courbe des J (h) pour v√©rifier qu'il s'agit bien du min
    plt.plot(h_aff, J_aff, color = col)
    plt.show()

    # On affiche l'estimateur optimal avec le h_opti
    x = np.linspace(start=-10, stop=10, num=200, endpoint="True")
    y = np.linspace(start=0, stop=len(x), num=len(x))

    i = 0
    while i < len(x):
        y[i] = f (h_opti, X, x[i], K_gaussien)
        i += 1

    plt.plot(x, y, color = col)
    plt.show()

trace_f_avec_h_optimise (X, "blue")

## Exercice 2

fleches = (
2, 1, 0, 4, 1, 0, 3, 2, 2, 3, 3, 2, 4, 3, 2, 5, 4, 2, 3, 2, 0, 3, 0, 0, 2, 2, 1, 4, 4, 1, 5, 4, 4, 4, 1, 0, 5, 3, 1, 5,
3, 2, 2, 0, 0, 3, 3, 1, 1, 1, 0, 3, 2, 1, 4, 4, 4, 4, 4, 2, 3, 3, 2, 2, 0, 0, 4, 2, 2, 3, 2, 0)

def trace_histo_f_avec_h_optimise (X, col) :
    plt.hist(X, normed=1)
    plt.show()

    trace_f_avec_h_optimise (fleches, col)

trace_histo_f_avec_h_optimise (fleches, "red")

# Ici les donn√©es sont issues de iris
from sklearn import datasets

# X correspond aux longueurs des s√©pales mesur√©s sur des fleurs
iris = datasets.load_iris()
X = iris.data[:, 1]  # we only take the first two features.
print (X)

trace_histo_f_avec_h_optimise (X, "red")
```
